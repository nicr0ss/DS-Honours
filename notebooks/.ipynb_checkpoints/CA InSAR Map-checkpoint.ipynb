```python
import scipy.io
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import contextily as ctx
from pykrige.ok import OrdinaryKriging
```

```python
SGMA = "../data/GWBasins.shp"
gdf = gpd.read_file(SGMA)
gdf.set_crs(epsg=3857, inplace=True)
gdf.set_index('OBJECTID', inplace=True)
gdf['Basin_Prefix'] = gdf['Basin_Numb'].str.split('-').str[0].astype(int)
```

```python
df = pd.read_csv("../data/CALIFORNIA_DWR_VERT_05.csv")
```

```python
geometry = [Point(xy) for xy in zip(df['LON'], df['LAT'])]
gdf_points = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

gdf = gdf.to_crs("EPSG:4326")
gdf_joined = gpd.sjoin(gdf_points, gdf, how="left", predicate="within")
df['Basin_Subb'] = gdf_joined['Basin_Subb']

df['Basin_Subb'] = df['Basin_Subb'].fillna('Unknown')
```

```python
sar_5021 = df[df.Basin_Subb.str.startswith("5-021")].copy()
```

```python
sar_5021
```

```python
sar_5021_gdf = gpd.GeoDataFrame(sar_5021, geometry=[Point(xy) for xy in zip(sar_5021['LON'], sar_5021['LAT'])])
```

```python
sar_5021_gdf.set_crs(epsg=4326, inplace=True)
```

```python
sar_5021_gdf.plot(figsize=(10, 6), marker='.', color='red', markersize=50)
plt.title('Cities on a Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
