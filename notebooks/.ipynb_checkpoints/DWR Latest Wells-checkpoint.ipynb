```python
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, DateFormatter
import seaborn as sns
import geopandas as gpd
import contextily as ctx
import numpy as np
from scipy.stats import gaussian_kde
import warnings
warnings.filterwarnings("ignore")
```

This notebook is to look at accessing the CA DWR ARCGIS well reporting data. The dashboard is [here](https://dwr.maps.arcgis.com/apps/dashboards/ba713497858845d0bf60d400f414d8c3). Primarily, the code below is used to access the data and transform it into a GeoPandas to assess how "live" the live wells really are. A separate notebook will look at how much historic data is available for each well and at what frequency.

```python
gis = GIS()
feature_layer_url = "https://services.arcgis.com/aa38u6OgfNoCkTJ6/ArcGIS/rest/services/GWL_Recently_Measured_v3/FeatureServer/2"
feature_layer = FeatureLayer(feature_layer_url, gis=gis)

query_result = feature_layer.query(where='1=1', out_fields='*')
sdf = query_result.sdf

sdf.head()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(15, 6))
sns.histplot(data=sdf, x='LAST_MSMT_DATE', bins=30, kde=True, color='skyblue')

ax = plt.gca()

ax.xaxis.set_major_locator(MonthLocator(interval=3))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))

plt.xticks(rotation=0)

plt.title('Histogram of LAST_MSMT_DATE', fontsize=16)
plt.xlabel('Measurement Date', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.histplot(sdf.dropna(subset=["WELL_DEPTH"])['WELL_DEPTH'])

plt.title('Histogram of WELL_DEPTH', fontsize=16)
plt.xlabel('Well Depth', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
gdf = gpd.GeoDataFrame(sdf, geometry='SHAPE')
```

```python
gdf['WELL_DEPTH'] = pd.to_numeric(gdf['WELL_DEPTH'], errors='coerce')
```

```python
gdf.set_crs(epsg=3857, inplace=True)
x = gdf.geometry.x
y = gdf.geometry.y

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Plot the original map with all points on the first subplot
gdf.plot(ax=ax1, color='blue', edgecolor='k', alpha=0.5, markersize=10)
ctx.add_basemap(ax1, crs=gdf.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax1.set_title("Map of California with Wells")

xy = np.vstack([x, y])
kde = gaussian_kde(xy)(xy)

sns.kdeplot(x=x, y=y, ax=ax2, fill=True, cmap="Reds", alpha=0.5, levels=100, thresh=0)
ctx.add_basemap(ax2, crs=gdf.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax2.set_title("Map of California with Heatmap of Well Spatial Density")

norm = plt.Normalize(vmin=kde.min(), vmax=kde.max())
sm = plt.cm.ScalarMappable(cmap="Reds", norm=norm)
sm.set_array([])

cbar = fig.colorbar(sm, ax=ax2, orientation='vertical', fraction=0.046, pad=0.04)
cbar.set_label('Density')

plt.show()
```

Above shows both points for each reporting location, and a heat map to look at where we see many wells in a small area.

```python
gdf['timestamp'] = gdf['LAST_MSMT_DATE'].astype(np.int64) // 10**9  # Convert to seconds since epoch
timestamps = gdf['timestamp']
```

```python
today_timestamp = datetime.now().timestamp()
gdf['days_ago'] = (today_timestamp - gdf['LAST_MSMT_DATE'].astype(np.int64) // 10**9) / (60 * 60 * 24)  # Convert seconds to days

# Define a 10-mile radius in meters (Web Mercator uses meters as units)
radius = 10 * 1609.34

def calculate_average_days_ago(point, gdf, radius):
    nearby_points = gdf[gdf.geometry.within(point.buffer(radius))]
    if len(nearby_points) > 0:
        return nearby_points['days_ago'].mean()
    else:
        return np.nan

gdf['average_days_ago'] = gdf.apply(lambda row: calculate_average_days_ago(row['SHAPE'], gdf, radius), axis=1)

norm = plt.Normalize(vmin=gdf['average_days_ago'].min(), vmax=gdf['average_days_ago'].max())

fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(column='average_days_ago', cmap="viridis", norm=norm, ax=ax, legend=False)  # Disable automatic legend
ctx.add_basemap(ax, crs=gdf.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_title("Heatmap of California with 10-Mile Average 'Days Ago'")

sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='vertical', fraction=0.046, pad=0.04)
cbar.set_label('Average Days Ago')

plt.show()
```

This plot takes the 10-mile average age of reportings for each station showing on the DWR feed. We an apparent spatial trend in recency of data shown by clusters of averages having similar ages.
