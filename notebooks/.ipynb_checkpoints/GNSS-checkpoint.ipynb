```python
import pandas as pd
import requests
import folium
import reverse_geocoder as rg
```

```python
stations = pd.read_csv("../data/GNSS_loc.csv")
```

```python
stations
```

```python
coordinates = list(zip(stations["Latitude (deg)"], stations["Longitude (deg)"]))

# Perform reverse geocoding
results = rg.search(coordinates)

# Extract the state from the results and add it as a new column
stations["State"] = [result["admin1"] for result in results]
```

```python
stations
```

```python
california_map = folium.Map(location=[36.7783, -119.4179], zoom_start=7)  # Center on California

# Add markers for each station
for index, row in stations[stations.State == "California"].iterrows():
    folium.Marker(
        location=[row["Latitude (deg)"], row["Longitude (deg)"]],
        popup=f"Station: {row['Station Name']}<br>ID: {row['ID']}<br>Elevation: {row['Ellipsoidal Elevation (m)']} m",
        tooltip=row["Station Name"]
    ).add_to(california_map)

# Display the map
california_map
```

```python
idaho_map = folium.Map(location=[44.0682, -114.7420], zoom_start=7)  # Center on Idaho

# Add markers for each station
for index, row in stations[stations.State == "Idaho"].iterrows():
    folium.Marker(
        location=[row["Latitude (deg)"], row["Longitude (deg)"]],
        popup=f"Station: {row['Station Name']}<br>ID: {row['ID']}<br>Elevation: {row['Ellipsoidal Elevation (m)']} m",
        tooltip=row["Station Name"]
    ).add_to(idaho_map)

# Display the map
idaho_map
```
