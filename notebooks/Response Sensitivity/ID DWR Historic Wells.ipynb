# Idaho DWR Well Data

Claim by the state that they have some of the best water reporting in the country. No direct data plugin found, data downloaded Nov 2024 located on an ArcGis page [here](https://www.arcgis.com/home/item.html?id=f1a190a2077c4b7da87b9cc19d0a316e). CSVs located in the _data_ folder and described in the readme.

```python
continuous = pd.read_csv("../data/Continuous_20240624.csv")
metadata = pd.read_csv("../data/Metadata_20240624.csv")
continuous["DateTime"] = pd.to_datetime(continuous["DateTime"])
```

```python
stations = continuous["Station Name"].unique()
summary_stats = pd.DataFrame(columns=["station", "earliest", "latest", "observations", "obs_gap"])


for station in tqdm(stations):
    filtered = continuous[continuous["Station Name"] == station]
    early = filtered.DateTime.min()
    late = filtered.DateTime.max()
    obs = len(filtered)
    gaps = (late - early).days / obs
    summary_stats = pd.concat([summary_stats,
                              pd.DataFrame([{"station":station, "earliest":early, "latest":late, "observations":obs,
                                            "obs_gap":gaps}])], ignore_index=True)
    
summary_stats = summary_stats[summary_stats["obs_gap"] > 0].copy()
```

```python
summary_stats
```

```python
summary_stats = summary_stats.merge(metadata, left_on="station", right_on="Station Name")
```

```python
summary_gpd = gpd.GeoDataFrame(summary_stats, 
                                geometry=gpd.points_from_xy(summary_stats['Longitude'], summary_stats['Latitude']),
                                crs="EPSG:4326")
summary_gpd.to_crs(epsg=3857, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
summary_gpd.plot(markersize=5, figsize=(10, 8), color="red", alpha=0.4, ax=ax)
ctx.add_basemap(ax, crs=summary_gpd.crs, source=ctx.providers.OpenStreetMap.Mapnik)
plt.title("Well Locations in Idaho")
plt.show()
```

This matches with the ID DWR portal located [here](https://idwr-groundwater-data.idaho.gov/applications/public.html?publicuser=public#waterdata/stationoverview).

### Same 2010 to 2020 logic as used on CA DWR periodic data

```python
buffer_distance = 10 * 1609.34

def meets_2010_2020(row, gdf):
    buffer = row.geometry.buffer(buffer_distance)
    neighbours = gdf[(gdf.geometry.within(buffer)) & (gdf.obs_gap < 365)]
    year_range = range(2010, 2021)
    for year in year_range:
        filtered = neighbours[(neighbours.earliest.dt.year <= year) | (neighbours.latest.dt.year >= year)]
        if len(filtered) == 0:
            return False
    return True
```

```python
summary_gpd["2010_2020?"] = summary_gpd.apply(lambda row: meets_2010_2020(row, summary_gpd), axis=1)
```

```python
len(summary_gpd[summary_gpd["2010_2020?"]]), len(summary_gpd)
```

Only one site does not make the cluster criteria.

```python
def find_10mi_avg_obs_sub365_2010_2020(row, gdf):
    if not row["2010_2020?"]:
        return -1
    buffer = row.geometry.buffer(buffer_distance)
    neighbours = gdf[(gdf.geometry.within(buffer)) & (gdf.obs_gap < 365)]
    total = 0
    year_range = range(2010, 2021)
    for year in year_range:
        filtered = neighbours[(neighbours.earliest.dt.year <= year) | (neighbours.latest.dt.year >= year)]
        expected_obs = sum(365 / filtered.obs_gap)
        total += expected_obs
    return total / 11
```

```python
summary_gpd["10mi_E(obs_1y)_2010_2020"] = summary_gpd.apply(
    lambda row: find_10mi_avg_obs_sub365_2010_2020(row, summary_gpd), axis=1)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
summary_gpd[summary_gpd["2010_2020?"]].plot(column="10mi_E(obs_1y)_2010_2020", legend=True, ax=ax, cmap='viridis')
ctx.add_basemap(ax, crs=summary_gpd.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_title(f"Map of 10-mile expected annual observations between 2010 and 2020")
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
summary_gpd[(summary_gpd["2010_2020?"])
            & (summary_gpd["10mi_E(obs_1y)_2010_2020"] < summary_gpd["10mi_E(obs_1y)_2010_2020"].quantile(0.975))
            & (summary_gpd["10mi_E(obs_1y)_2010_2020"] > summary_gpd["10mi_E(obs_1y)_2010_2020"].quantile(0.025))
            ].plot(column="10mi_E(obs_1y)_2010_2020", legend=True, ax=ax, cmap='viridis') # trimmed for middle 95% of data
ctx.add_basemap(ax, crs=summary_gpd.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_title(f"Map of 10-mile expected annual observations between 2010 and 2020, middle 95%")
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
summary_gpd[(summary_gpd["2010_2020?"])
            & (summary_gpd["10mi_E(obs_1y)_2010_2020"] < summary_gpd["10mi_E(obs_1y)_2010_2020"].quantile(0.2))
            ].plot(column="10mi_E(obs_1y)_2010_2020", legend=True, ax=ax, cmap='viridis') # trimmed for middle 95% of data
ctx.add_basemap(ax, crs=summary_gpd.crs, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_title(f"Map of 10-mile expected annual observations between 2010 and 2020, bottom 20%")
plt.axis('off')
plt.show()
```

Idaho seems to have good temporal resolution with plenty of observations available over long time spans, especially when clustered into 10mi regions. However, their spatial resolution is not as strong as CA.
